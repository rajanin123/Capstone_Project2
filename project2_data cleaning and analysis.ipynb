MERGING DATASETS

```python
# Merging data set sales and customers
df1 = pd.merge(df_Sa, df_cu, on = 'CustomerKey', how = 'inner')
df1.drop_duplicates(inplace = True)
df1.dropna(inplace = True)
df1 = df1.drop(columns=['Delivery_Date'])
df1.head()
```

```python
# Merge dataset df1(merged sales and customers data) with Exchange_rates
df2 = pd.merge(df1, df_Ex, left_on = ['Order_Date', 'Currency_Code'], right_on = ['Date', 'Currency'], how = 'inner')
df2.head()
```

```python
# merge df2 (df1(sales+customers+Exchange_rates)) with  products
df3 = pd.merge(df2, df_Pr, on = 'ProductKey', how = 'inner')
df3.head()
```

```python
overall = df3
overall.head()
```

```python
df1_overall = pd.merge(overall, df_st ,on=['StoreKey','Country','State'],how='inner')
```

Create Table Stores in Postgresql

```python
# Define PostgreSQL database connection details and create 'stores' data table
db_config = {
    'host': 'localhost',
    'dbname': 'capstone_project2',
    'user': 'postgres',
    'password': 'agcharan',
    'port': 5432}

# Establish connection to PostgreSQL

conn = psycopg2.connect(**db_config)
cursor = conn.cursor()
print("Connected to PostgreSQL")

# Create a table stores
create_table_query = '''create table if not exists stores(StoreKey int,
                                                        Country varchar(100),
                                                        State varchar(100),
                                                        Square_Meters float,
                                                        Open_Date date
                                                        );'''
cursor.execute(create_table_query)
conn.commit()

print("Table created successfully")


```

```python
# Define the SQL insert query (imserting CSV data from 'df_st' dataset and create 'stores' Table)
insert_query = '''
INSERT INTO stores (StoreKey, Country, State, Square_Meters, Open_Date )
VALUES (%s, %s, %s, %s, %s);
'''

# Loop through DataFrame and insert data into PostgreSQL
try:
    for index, row in df_st.iterrows():
        cursor.execute(insert_query, (row['StoreKey'], row['Country'], row['State'], row['Square_Meters'], row['Open_Date']))
    conn.commit()
    print("CSV data inserted successfully")
except Exception as e:
    conn.rollback()
    print(f"Error inserting data: {e}")
```
